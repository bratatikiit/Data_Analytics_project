# 1. Code Purpose:

**This Python code creates a box plot to visualize the distribution of total vaccinations across different countries and years**, using data from a CSV file.

## Code Explanation: ##

* 1. Import Libraries:

pandas (pd): Used for data manipulation tasks like reading CSV files, grouping data, and transforming data structures.
matplotlib.pyplot (plt): Used for creating visualizations like box plots in this case.
* 2. Read CSV Data:

df = pd.read_csv("./vaccinations-by-manufacturer.csv"):
Reads the CSV file located at "./vaccinations-by-manufacturer.csv".
Stores the data in a pandas DataFrame named df.
Assumptions: The code assumes the CSV file has the following columns:
"Country Name" (string): Name of the country.
"Year" (numeric): Year of vaccination data.
"total_vaccinations" (numeric): Number of vaccinations administered.
* 3. Group and Unstack Data:

grouped_data = df.groupby(["Country Name", "Year"])["total_vaccinations"].sum():
Groups the data in the DataFrame df by two columns:
"Country Name": Groups data for each country.
"Year": Groups data for each year.
Calculates the sum of "total_vaccinations" for each group (country-year combination).
This creates a Series object with multi-level indexing, where the first level is the country name and the second level is the year.
unstacked_data = grouped_data.unstack():

Transforms the Series object grouped_data into a DataFrame named unstacked_data.
Unstacks the multi-level index, creating columns for each year and rows for each country.
This makes the data more suitable for creating a box plot, where each box represents the distribution of vaccination counts for a specific country across different years.
* 4. Create the Box Plot:
plt.figure(figsize=(10, 6)):
Creates a figure object for the plot, setting its size (width and height in inches) to 10 and 6, respectively. You can adjust these values based on your preference.
unstacked_data.plot(kind="box", vert=False):
Creates a box plot using the DataFrame unstacked_data.
The kind="box" argument specifies a box plot visualization.
The vert=False argument sets the boxes to be oriented horizontally, which can be more readable when dealing with many countries.
* 5. Customize the Plot:

plt.xlabel("Total Vaccinations"): Sets the label for the x-axis, indicating it represents the total number of vaccinations.
plt.ylabel("Year"): Sets the label for the y-axis, indicating it represents the year.
plt.title("Distribution of Total Vaccinations by Country and Year (2020-2024)"): Sets the title of the plot, providing context for the data being visualized.
plt.xticks(rotation=45): Rotates the x-axis labels by 45 degrees, improving readability when there are many countries and their labels might overlap.
plt.tight_layout(): Adjusts the spacing between plot elements (title, labels, ticks, etc.) for a cleaner presentation.
* 6. Display the Plot:
plt.show(): Displays the created box plot in the Jupyter Lab environment .

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV data into a DataFrame
df = pd.read_csv("./vaccinations-by-manufaturer.csv")
print(df.head(10))
# Group data by country and year ( Year is numeric)
grouped_data = df.groupby(["Country Name", "Year"])["total_vaccinations"].sum()
unstacked_data = grouped_data.unstack()  # Unstack for box plot

# Create the box plot
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
unstacked_data.plot(kind="box", vert=False)  # Create box plot with horizontal boxes

# Customize the plot
plt.xlabel("Total Vaccinations")
plt.ylabel("year")
plt.title("Distribution of Total Vaccinations by Country and Year (2020-2024)")
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()

# Display the plot in Jupyter Lab
plt.show()
```

csv_file = './vaccinations-by-manufaturer.csv': Defines the path to the CSV file assumed to be located in the same directory as the script.

year_col = 'Year' (replace with actual name): Stores the column name containing year information in the CSV file.

total_vaccinations_col = 'total_vaccinations' (assuming a total vaccinations column): Stores the column name containing total vaccination data.

vaccine_totals = df.groupby([year_col, vaccine_col])[total_vaccinations_col].sum().unstack():

Groups the data in the DataFrame df by two columns:
year_col: Groups data for each year (as defined by year_col).
vaccine_col: Groups data for each vaccine type (as defined by vaccine_col).
Calculates the sum of total_vaccinations_col (assumed to be total vaccinations) for each group (year-vaccine type combination).

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV file is named 'vaccinations.csv' and is in the same directory
csv_file = './vaccinations-by-manufaturer.csv'

# Error handling for potential CSV loading issues
try:
    df = pd.read_csv(csv_file)
except FileNotFoundError:
    print(f"Error: CSV file '{csv_file}' not found. Please check the file path.")
    exit()

# Extract year and vaccine columns from your data (replace with actual names if different)
year_col = 'Year'
vaccine_col = 'vaccine_type'
total_vaccinations_col = 'total_vaccinations'  # Assuming a total vaccinations column

# Group data by year and vaccine type, summing total vaccinations
vaccine_totals = df.groupby([year_col, vaccine_col])[total_vaccinations_col].sum().unstack()

# Create the plot (stacked bar chart)
vaccine_totals.plot(kind="bar", stacked=True, colormap="tab20")
plt.xlabel(year_col)  # Set appropriate label based on your column name
plt.ylabel("Total Vaccinations")
plt.title("Total Vaccinations by Vaccine Type (2020-2024)")  # Adjust title based on your data range
plt.xticks(rotation=0)  # Rotate x-axis labels for better readability
plt.legend(title="Vaccine Type")
plt.grid(axis="y")  # Add grid lines for better visual reference

plt.tight_layout()  # Adjust spacing to prevent overlapping elements
plt.show()
```

data = pd.read_csv('./vaccinations-by-age-group.csv'):

data['date'] = pd.to_datetime(data['date'], format='%d/%m/%Y'):
This line converts the 'date' column (assuming it exists in the CSV) to a datetime format. This can be helpful for time-series analysis but might not be strictly necessary for correlation calculations.

Defines a list of column names containing numeric data for which correlations will be calculated. You might need to adjust this list based on your actual column names.

5. Create the Heatmap: plt.figure(figsize=(10, 6)):

Creates a figure object for the plot, setting its size (width and height in inches) to 10 and 6, respectively. You can adjust these values based on your preference.
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm'):

**Creates a heatmap using the seaborn library.**
correlation_matrix is the data source for the heatmap, containing the correlation coefficients.
annot=True displays the correlation values within each cell of the heatmap for better readability.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


data = pd.read_csv('./vaccinations-by-age-group.csv')

# Convert the 'date' column to datetime format (optional, but recommended for clarity)
data['date'] = pd.to_datetime(data['date'], format='%d/%m/%Y')

# Select the numeric columns for correlation analysis
numeric_columns = ['people_vaccinated_per_hundred', 'people_fully_vaccinated_per_hundred', 'people_with_booster_per_hundred']
correlation_matrix = data[numeric_columns].corr()

# Generate the heatmap
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap (Vaccinated, Fully Vaccinated, With Booster)')
plt.show()
```

# * 14. This Python code creates a heatmap visualization using Seaborn to represent the distribution of affected cases by month and year . 

It reads data from a CSV file, preprocesses it for the heatmap, and displays the visualization with informative labels and a title.

* 1. Import Libraries:

pandas (pd): Used for data manipulation tasks like reading CSV files, grouping data, and pivoting DataFrames.
seaborn (sns): Used for creating statistical plots with enhanced aesthetics compared to Matplotlib's default plots.
matplotlib.pyplot (plt) (imported but not explicitly used): Although not directly used in this code, it's necessary for Seaborn to function as it builds upon Matplotlib for visualization.
* 2. Read CSV Data:

df = pd.read_csv('./affected_country.csv'): Reads the CSV file located at './affected_country.csv' and stores the data in a pandas DataFrame named df.
* 3. Handle Duplicate Entries (Optional):

df_aggregated = df.groupby(['month', 'year'], as_index=False)['affected'].sum(): This line assumes there might be duplicate entries in the DataFrame for the same month and year with "affected" values. It groups the data by "month" and "year" columns, calculates the sum of "affected" cases for each group, and stores the result in a new DataFrame named df_aggregated. You can comment out this line if you're certain there are no duplicates.
* 4. Pivot Data for Heatmap:

heatmap_data = df_aggregated.pivot(index='month', columns='year', values='affected'): This line creates a new DataFrame named heatmap_data specifically designed for the heatmap. It pivots the df_aggregated DataFrame, using "month" as the index (rows), "year" as the columns, and keeps the "affected" values in the data cells. This restructuring is necessary for the heatmap to represent the data by month and year.
* 5. Create the Heatmap:

plt.figure(figsize=(10, 8)): Creates a new figure window for the plot with a width of 10 inches and a height of 8 inches, providing a clear canvas for the heatmap.
sns.heatmap(heatmap_data, annot=True, fmt="f", cmap="YlGnBu"): 
Creates a heatmap using Seaborn's heatmap function with the following arguments:
heatmap_data: The DataFrame (heatmap_data) containing the data for the heatmap.
annot=True: Enables annotation of data values on top of each heatmap cell.
fmt="f": Formats the annotation values to display as floating-point numbers.
cmap="YlGnBu": Selects a colormap named "YlGnBu" for the heatmap, which typically ranges from yellow (low values) to green and blue (high values).
* 6. Customize the Plot:

plt.title("Heatmap of Affected Cases by Month and Year in India (2021)"): Sets a title for the plot, providing context for the data being visualized.
plt.xlabel("Year"): Sets a label for the x-axis, indicating that it represents the year.
plt.ylabel("Month"): Sets a label for the y-axis, indicating that it represents the month.
* 7. Display the Plot:
plt.show(): Displays the created heatmap.

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the CSV data into a DataFrame
df = pd.read_csv('./affected_country.csv')

# Ensure there are no duplicate entries
df_aggregated = df.groupby(['month', 'year'], as_index=False)['affected'].sum()

# Pivot the DataFrame to prepare for heatmap
heatmap_data = df_aggregated.pivot(index='month', columns='year', values='affected')

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data, annot=True, fmt="f", cmap="YlGnBu")

# Customize the plot
plt.title("Heatmap of Affected Cases by Month and Year in world")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```

# 15. This Python code creates a world map visualization using Cartopy to highlight countries affected based on your CSV data. 

It reads country data, filters affected countries from your DataFrame, colors them red on the map, and displays the map with coastlines and a title.

* 1. Import Libraries:

pandas (pd): Used for data manipulation tasks like reading CSV files and accessing data by column names.
matplotlib.pyplot (plt): Used for creating the basic plot structure for the map.
from cartopy.io import shapereader: Imports the shapereader function from Cartopy for reading country shapefile data.
from cartopy import crs: Imports coordinate reference system (CRS) functionalities from Cartopy for map projections.
* 2. Read CSV Data:
data = pd.read_csv("./affected_country.csv"): Reads the CSV file located at "./affected_country.csv" and stores the data in a pandas DataFrame named data.
* 3. Filter Affected Countries:

affected_countries = data["country"].unique(): Extracts a list of unique countries with affected cases from the "country" column in the DataFrame data.
* 4. Create the Figure and Map Axes:

plt.figure(figsize=(16, 6)): Creates a new figure window for the plot with a width of 16 inches and a height of 6 inches, providing a larger canvas for the map.
ax = plt.axes(projection=crs.PlateCarree()): Creates a map axes using the Plate Carree projection, which is commonly used for world maps.
* 5. Download and Load Country Data (High Resolution):

shpfile = shapereader.natural_earth(resolution='110m', category='cultural', name='admin_0_countries'): Downloads high-resolution country data (110 meters) from the Natural Earth database using Cartopy's shapereader. The cultural category ensures you get country borders (administrative boundaries).
reader = shapereader.Reader(shpfile): Creates a reader object to access the downloaded country data.
countries = reader.records(): Reads the country data from the shapefile and stores it in a list of country records.
* 6. Highlight Affected Countries:

for country in countries: Iterates over each country record in the countries list.
country_name = country.attributes['NAME']: Extracts the full country name from the record's attributes dictionary using the key "NAME" (replace "ADM0_A3" if your data uses a 3-letter country code).
if country_name in affected_countries: Checks if the current country name is present in the list of affected_countries.
ax.add_geometries(country.geometry, crs.PlateCarree(), facecolor='red', edgecolor='k'): If the country is affected, adds its geometry (borders) to the map axes with a red face color and black edge color.
else: If the country is not affected, adds its geometry with a white face color and black edge color.
* 7. Add Map Elements:

ax.coastlines(color='black', linewidth=0.6): Adds coastlines to the map with a black color and a line width of 0.6 for better visualization.
* 8. Display the Plot:
plt.title("World Map of Affected Countries"): Sets a title for the plot, indicating the map's purpose.
plt.show(): Displays the created world map.

```python
import pandas as pd
import matplotlib.pyplot as plt
from cartopy.io import shapereader
from cartopy import crs

# Read the CSV data
data = pd.read_csv("./affected_country.csv")

# Filter for unique countries with affected cases
affected_countries = data["country"].unique()

# Create a figure with a specific size
plt.figure(figsize=(16, 6))
ax = plt.axes(projection=crs.PlateCarree())

# Download and load country data (high resolution)
shpfile = shapereader.natural_earth(resolution='110m', category='cultural', name='admin_0_countries')
reader = shapereader.Reader(shpfile)
countries = reader.records()

# Highlight affected countries in red
for country in countries:
    country_name = country.attributes['NAME']  # Use 'NAME' instead of 'ADM0_A3' for full country names
    if country_name in affected_countries:
        ax.add_geometries(country.geometry, crs.PlateCarree(), facecolor='red', edgecolor='k')
    else:
        ax.add_geometries(country.geometry, crs.PlateCarree(), facecolor='white', edgecolor='k')

# Add additional map elements
ax.coastlines(color='black', linewidth=0.6)  # Add coastlines

# Display the plot
plt.title("World Map of Affected Countries")
plt.show()
```
